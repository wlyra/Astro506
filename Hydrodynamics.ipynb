# Hydrodynamics

## Distribution function 


To describe the state of the gas statistically, we define the distribution function $f (\mathbf{x}, \mathbf{v}, t)$ such that $f d^3x d^3v$ is the average number of particles contained in a volume element $d^3x$ about $\mathbf{x}$ and a velocity-space element $d^3v$ about $\mathbf{v}$ at time $t$. We demand that $f \geq 0$ and that $f \rightarrow 0$ as the components of the particle velocity tends to infinity, sufficiently rapidly to guarantee that a finite number of particles has a finite energy.

Macroscopic properties of the gas can be computed from $f (\mathbf{x}, \mathbf{v}, t)$, e.g.:

\begin{eqnarray}
n(\mathbf{x}, t) &=& \int_{-\infty}^\infty f (\mathbf{x}, \mathbf{v}, t) \ d^3v \\
\rho(\mathbf{x}, t) &=& m \ n(\mathbf{x}, t)\\
\mathbf{u}(\mathbf{x}, t) &=& \frac{1}{n}\int_{-\infty}^\infty f (\mathbf{x}, \mathbf{v}, t) \ \mathbf{v} \ d^3v = \langle \mathbf{v} \rangle 
\end{eqnarray}


i.e. the number density of the particles $n$, the mass density $\rho$ (where $m$ is the mass of a single particle) and the average velocity of an element of gas $\mathbf{u}$ (= macroscopic flow velocity). 

To study macroscopic properties of the gas it is useful to decompose the particle velocity $\mathbf{v}$ into the average velocity $\mathbf{u}$ and a random velocity relative to the mean flow $\mathbf{w}$


\begin{equation}
\mathbf{v} = \mathbf{u} + \mathbf{w}
\end{equation}


## Boltzmann Transport Equation

We now search for an equation describing the changes of $f (\mathbf{x}, \mathbf{v}, t)$ with time. Consider a group of particles located in a phase-space volume element $(d\mathbf{x}_0, d\mathbf{v}_0)$ around ($\mathbf{x}_0, \mathbf{v}_0$). We assume that an external force $\mathbf{F}(\mathbf{x}, t)$ is acting on the particles so that they experience an acceleration $\mathbf{a}(\mathbf{x}, t) = \mathbf{F}(\mathbf{x}, t)/m$. For the moment, we ignore collisions (and interactions) between the particles. The phase-space element will evolve into ($d\mathbf{x},d\mathbf{v}$) around ($\mathbf{x},\mathbf{v}$) where


\begin{eqnarray}
\mathbf{x} &=& \mathbf{x}_0 + \mathbf{v}_0 dt \\
\mathbf{v} &=& \mathbf{v}_0 + \mathbf{a} dt 
\end{eqnarray}

Because of the Hamiltonian nature of the coordinates $\mathbf{x},\mathbf{v}$, the transformation conserves volume in phase space (Liouville's theorem)

\begin{equation}
d^3x \ d^3v = d^3x_0 \ d^3v_0
\end{equation}

The number of particles inside the original phase-space element is

\begin{equation}
dN_0 =f(\mathbf{x}_0,\mathbf{v}_0,t_0)\ d^3x_0 \ d^3v_0 
\end{equation}


In the absence of collisions all these particles must end up in the new element and therefore 
$dN_0 = dN$ where

\begin{eqnarray}
dN &=& f(\mathbf{x},\mathbf{v},t) \ d^3x \ d^3v \\
&=& f(\mathbf{x}_0 +\mathbf{v}_0 dt,\mathbf{v}_0 +\mathbf{a}dt,t_0 +dt)\ d^3x \ d^3v
\end{eqnarray}

Together with Liouville's theorem, this implies that

\begin{equation}
f(\mathbf{x}_0 + \mathbf{v}_0dt,\mathbf{v}_0 +\mathbf{a}dt,t_0 + dt)=f(\mathbf{x}_0,\mathbf{v}_0,t_0)
\end{equation}

i.e. in the absence of collisions the phase-space density of a group of particles is invariant. By expanding to first order in $dt$ we find

\begin{equation}
\frac{\partial f}{\partial t} + \mathbf{v}\cdot\nabla_x f + \mathbf{a}\cdot\nabla_v f = 0 
\end{equation}


which is known as the <b>collisionless Boltzmann's equation</b>, or <b>Vlasov’s equation</b>.



Notice that the equation can be written as 

\begin{equation}
\frac{\partial f}{\partial t} + \nabla_{6D}\cdot \left( f \mathbf{W} \right) 
\end{equation}

where $\mathbf{W} = (\mathbf{x},\mathbf{v})$ is a 6D state vector and $\nabla_{6D}$ is a 6D divergence. This equation has the form of the continuity equation. Boltzmann's equation is derived under the condition that no particle can jump in phase space. Particles can only go smoothly (continuously) from one place to another. Thus, the Boltzmann equation is just the continuity equation of the flow of particles in phase space. 


### Collisions

When particles suffer a collision during $dt$ their velocities will be changed and they will not, in general, end up in the velocity element $d\mathbf{v}$ centered around $\mathbf{v}$ but in a different velocity element $d\mathbf{v}^\prime$ centered around $\mathbf{v}^\prime$ with $\mathbf{v}^\prime \neq \mathbf{v}$. On the other hand, particles not contained in the original element $d\mathbf{v}_0$ may end up in $d\mathbf{v}$ due to collisions. Therefore, $f$ is no longer invariant. To take these processes into account we have to add a term on the right-hand side of Vlasov's equation which gives the net rate at which particles are entering into the phase-space element under consideration. We write this term symbolically as

\begin{equation}
\left(\frac{Df}{Dt}\right)_{\rm coll} 
\end{equation}

and obtain the Boltzmann transport equation

\begin{equation}
\frac{\partial f}{\partial t} + \mathbf{v}\cdot\nabla_x f + \mathbf{a}\cdot\nabla_v f = \left(\frac{Df}{Dt}\right)_{\rm coll} 
\end{equation}

The right-hand side of Boltzmann's equation is defined as the net rate at which particles are entering into the phase-space element under consideration. Thus, we can write


\begin{equation}
\left(\frac{Df}{Dt}\right)_{\rm coll}  = R_{\rm in} - R_{\rm out}
\end{equation}


where $R_{\rm in}$ and $R_{\rm out}$ are the rates at which particles are scattered in and out of a given phase-space element due to collisions.

First, we calculate the rate at which particles of type 1 are scattered out of $d^3x_1 d^3v_1$. For each particle of type 1 the number of particles of type 2 moving with velocities in the range ($\mathbf{v}_2, \mathbf{v}_2 + d\mathbf{v}_2$), incident within a range of impact parameters (b, b + db) and within azimuth range $d\phi$ in a unit time is 

\begin{equation}
(f_2 \ d^3v_2) \ q \ b \ db \ d\phi
\end{equation}

where $q = |\mathbf{v}_2-\mathbf{v}_1|$ and $f_2 \equiv f(\mathbf{x},\mathbf{v}_2,t)$. The total number of collisions is given by integrating over all impact parameters, azimuths and incident velocities, and then multiplying by the number of particles of type 1, i.e. 

\begin{equation}
f_1 d^3x_1 d^3v_1 
\end{equation}

We obtain

\begin{equation}
R_{\rm out}  d^3x_1 d^3v_1 = \int f_1 f_2 \ q \ b \ db \ d\phi \ d^3v_2 d^3x_1 d^3v_1 
\end{equation}

Rin can be evaluated by considering the inverse enounters, i.e. collisions where initial velocities $\mathbf{v}_1^\prime$ and $\mathbf{v}_2^\prime$ are changed to the final values $\mathbf{v}_1$ and $\mathbf{v}_2$ within the considered phase-space element. We find

\begin{equation}
R_{\rm in} d^3x_1 d^3v_1  = \int f_1^\prime f_2^\prime \ q^\prime \ b \ db \ d\phi \ d^3v_2^\prime d^3x_1 d^3v_1^\prime 
\end{equation}

where $f_i^\prime \equiv f(\mathbf{x},\mathbf{v}_i^\prime,t)$. It can be shown that $q = q^\prime$ and that $d^3v_1 d^3v_2 = d^3v_1^\prime d^3v_2^\prime$. Therefore we can re-write $R_in$ as

\begin{equation}
R_{\rm in} d^3x_1 d^3v_1  = \int f_1^\prime f_2^\prime \ q \ b \ db \ d\phi \ d^3v_2 d^3x_1 d^3v_1^\prime 
\end{equation}

Thus we obtain

\begin{equation}
\left(\frac{Df}{Dt}\right)_{\rm coll} = \int \left(f_1^\prime f_2^\prime-f_1f_2 \right)q \ b \ db \ d\phi \ d^3v_2
\end{equation}

Introducing the cross section $\sigma(\Omega)d\Omega = bdbd\phi$ instead of the impact parameter and azimuth we can write the Boltzmann equation, accounting for binary collisions, as

\begin{equation}
\boxed{
\left(\frac{Df}{Dt}\right)_{\rm coll} = \int \left(f_1^\prime f_2^\prime-f_1f_2 \right)q \sigma(\Omega)d\Omega \ d^3v_2
}
\end{equation}



#### Properties of the Collision Integral

Let $Q(\mathbf{v})$ be any function of the particle velocity $\mathbf{v}$ and define

\begin{eqnarray}
I &\equiv& \int Q(\mathbf{v}_1) \left(f_1^\prime f_2^\prime-f_1f_2 \right)q \sigma(\Omega)d\Omega \ d^3v_1\ d^3v_2 \\
&=& \int Q(\mathbf{v}_1)  \left(\frac{Df}{Dt}\right)_{\rm coll}\ d^3v_1
\end{eqnarray}


If we just interchange the labeling 1 and 2 of the particles then obviously the value of $I$ does not change. Adding these two expressions we get
 
\begin{equation}
I = \frac{1}{2}\int \left[Q(\mathbf{v}_1)-Q(\mathbf{v}_2)\right] \left(f_1^\prime f_2^\prime-f_1f_2 \right)q \sigma(\Omega)d\Omega \ d^3v_1\ d^3v_2 
\end{equation}
 
 
If we replace the collision by its inverse the integral still has the same value because for every collision exist an inverse collision with the same cross section. Therefore we get

\begin{equation}
I = \frac{1}{2}\int \left[Q(\mathbf{v}_1^\prime)-Q(\mathbf{v}_2^\prime)\right] \left(f_1f_2-f_1^\prime f_2^\prime \right)q \sigma(\Omega)d\Omega \ d^3v_1 \ d^3v_2 
\end{equation}

 
Adding the two expressions

\begin{equation}
I = -\frac{1}{4}\int \left[\delta Q(\mathbf{v}_1)+\delta Q(\mathbf{v}_2)\right] \left(f_1^\prime f_2^\prime -f_1f_2 \right)q \sigma(\Omega)d\Omega \ d^3v_1 \ d^3v_2 
\end{equation}

where we have defined $\delta Q(\mathbf{v}) = Q(\mathbf{v}^\prime)-Q(\mathbf{v})$. 
 
By definition

\begin{equation}
\delta Q(\mathbf{v}_1)+ \delta Q(\mathbf{v}_2) = \left[Q(\mathbf{v}_1^\prime)+ Q(\mathbf{v}_2^\prime)\right]  -\left[Q(\mathbf{v}_1)+ Q(\mathbf{v}_2)\right] 
\end{equation}

 
Thus, if $Q$ is a quantity which is conserved during the collision (mass, momentum, energy), then $\delta Q(\mathbf{v}_1)+ \delta Q(\mathbf{v}_2) = 0$, i.e. the integrand in the collision integral vanishes so consequently $I = 0$. We will use this property when deriving the conservation laws of a fluid from the Boltzmann equation.


#### Moments of the Boltzmann Equation

The equations of fluid dynamics can be derived by calculating moments of the Boltzmann equation for quantities that are conserved in collisions of the particles.

##### The Conservation Theorem

We form moments of the Boltzmann equation by multiplying by a quantity $Q$ and integrating 

\begin{equation}
\int Q \left( \frac{\partial f}{\partial t} + \mathbf{v}\cdot\nabla_x f + \mathbf{a}\cdot\nabla_v f \right) d^3 v_1  = \int Q \left(\frac{Df}{Dt}\right)_{\rm coll} d^3 v_1 
\end{equation}

If $Q$ is conserved in the collision, the RHS cancels, and the equation reduces to a conservation law. In the following discussion we assume that $Q$ is conserved. The momentum equation then can be transformed into the conservation theorem. 

\begin{equation}
\frac{\partial }{\partial t} \left(n\langle Q \rangle\right) + \nabla_x \cdot \left(n\langle Q  \mathbf{v} \rangle\right) - n\mathbf{a}\cdot \langle \nabla_v Q \rangle = 0 
\end{equation}

where we defined the average value $\langle A \rangle$ of any quantity $A$ as

\begin{equation}
\langle A \rangle = \frac{\int A f d^3v}{\int f d^3v} = \frac{1}{n}\int A f d^3v
\end{equation}

For a gas consisting of particles which have no inner structure we have 5 conserved
quantities $Q$, these are the mass $m$, the three components of the momentum $m\mathbf{v}$, and the energy $mv^2/2$.


##### The Equation of Continuity

If we choose $Q = m$ and insert this into the conservation equation, we get

\begin{equation}
\frac{\partial }{\partial t} \left(n m\right) + \nabla_x \cdot \left(nm\langle \mathbf{v} \rangle \right)  = 0 
\end{equation}

Using $nm=\rho$ and $\mathbf{u} = \langle \mathbf{v} \rangle$ we obtain the <b>continuity equation</b>

\begin{equation}
\boxed{
\frac{\partial \rho}{\partial t}  + \nabla \cdot \left(\rho \mathbf{u} \right)  = 0 
}
\end{equation}


##### The Momentum Equation

If $Q = mv_i$ (the i-th component of the particle momentum), the conservation equation becomes


\begin{equation}
\frac{\partial }{\partial t} \left(n m \langle v_i \rangle\right) + \frac{\partial}{\partial x_j}\left(nm\langle v_jv_i \rangle\right) - nma_j \delta_{ij} = 0 
\end{equation}

Using the decomposition $v_i = u_i + w_i$ (particle velocity = mean flow velocity + random component) and keeping in mind that $\langle w_i \rangle = 0$ and $\langle v_i \rangle \equiv u_i$ we obtain the momentum equation

##### The Navier-Stokes Equation

If $Q = mv_i$ (the i-th component of the particle momentum), the conservation equation becomes


\begin{equation}
\frac{\partial }{\partial t} \left(n m \langle v_i \rangle\right) + \frac{\partial}{\partial x_j}\left(nm\langle v_jv_i \rangle\right) - nma_j \delta_{ij} = 0 
\end{equation}

Using the decomposition $v_i = u_i + w_i$ (particle velocity = mean flow velocity + random component) and keeping in mind that $\langle w_i \rangle = 0$ and $\langle v_i \rangle \equiv u_i$ we obtain the momentum equation (aka Navier-Stokes equation)

\begin{equation}
\boxed{
\frac{\partial }{\partial t} \left(\rho u_i\right) + \frac{\partial}{\partial x_j}\left(\rho u_ju_i   + P\delta_{ij} - \Pi_{ij}\right) = \rho a_i
}
\end{equation}

Here we have used

\begin{equation}
\rho\langle w_i w_j \rangle = P\delta_{ij} - \Pi_{ij}  
\end{equation}

where 

\begin{equation}
P \equiv \frac{1}{3} \rho \langle w^2 \rangle
\end{equation}

is the gas pressure, and


\begin{equation}
\Pi_{ij}  \equiv \rho \left(\frac{1}{3} \langle w^2 \rangle\delta_{ij} - \langle w_i w_j \rangle\right)
\end{equation}

is the viscous stress tensor.


##### The Energy Equation

If we choose $Q = mv^2/2$, the conservation equation assumes the form


\begin{equation}
\frac{\partial }{\partial t} \left(\frac{1}{2}n m \langle v^2 \rangle\right) + \frac{\partial}{\partial x_i}\left(nm\langle v^2v_i \rangle\right) - nma_i \langle v_i \rangle = 0 
\end{equation}

Using the same decomposition of $v_i$ and the same definitions as before we obtain the energy equation

\begin{equation}
\frac{\partial }{\partial t} \left(\frac{1}{2}\rho u^2 + \rho e \right) + \frac{\partial}{\partial x_i}\left[ \left(\frac{1}{2}\rho u^2 + \rho e \right)u_i + \left(P\delta_{ij} - \Pi_{ij}\right)u_j + C_i\right] = \rho a_i u_i 
\end{equation}

where 


\begin{equation}
e \equiv \frac{1}{2} \langle w^2\rangle
\end{equation}

is the specific internal energy (per mass), and

\begin{equation}
C \equiv \frac{\rho}{2} \langle w^2\mathbf{w}\rangle
\end{equation}

is the conduction heat flux. These conservation equations are exact for the adopted model of the gas but have no practical value until we can evaluate $\Pi_{ij}$ and $C_i$. Using kinetic theory this can be done from first principles.


##### Conservation Equations for Equilibrium Flow

The gas can be considered to be in local equilibrium if particle mean free paths are very small compared to characteristic length scales of the flow and if gradients are sufficiently small. Then we can assume that the distribution function $f(\mathbf{x},\mathbf{v},t)$ is given by a Maxwellian velocity distribution

\begin{equation}
f(w) =  n \left(\frac{m}{2\pi k T}\right)^{3/2} \mathrm{exp}\left(-\frac{mw^2}{2kT}\right)
\end{equation}

with a local temperature $T$ and particle density $n$ ($w$ is the random part of the particle velocity as before).

Using this distribution function it can be shown that $\Pi_{ij}$ and $C_i$ as defined in the previous section are <b>zero</b>. Therefore, the conservation equations become

\begin{eqnarray}
\frac{\partial \rho}{\partial t}  + \nabla \cdot \left(\rho \mathbf{u} \right)  &=& 0 \\
\frac{\partial }{\partial t} \left(\rho u_i\right) + \frac{\partial}{\partial x_j}\left(\rho u_ju_i   + P\delta_{ij} \right) &=& \rho a_i\\
\frac{\partial }{\partial t} \left(\frac{1}{2}\rho u^2 + \rho e \right) + \frac{\partial}{\partial x_i}\left[ \left(\frac{1}{2}\rho u^2 + \rho e \right)u_i + P\delta_{ij}u_i\right] &=& \rho a_i u_i 
\end{eqnarray}
